```python
#figure
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

fig = plt.figure(figsize = (12, 4.5))
axes = []
axes.append(fig.add_subplot(1, 2, 1))
axes.append(fig.add_subplot(1, 2, 2))

# 用 for 循环，在两张画纸上分别绘制多组散点图
for i in range(2):
    for category, feature in data_dict_list[i].items(): # 获取对应的数据字典
        axes[i].scatter(feature['平均票价'],               # 获取'平均票价'列表作为 x 轴坐标 
                        feature['平均上座'],               # 获取'平均上座'列表作为 y 轴坐标
                        marker=marker_dict[category],        # 从准备好的 marker_dict 设置数据形状
                        label=category          # 设置数据标签
                       )

    # 设置子图标题
    axes[i].set_title(title_list[i], fontsize=16)
    
    # 设置坐标轴标签、网格、图例
    axes[i].set_xlabel('平均票价（元）')
    axes[i].set_ylabel('平均上座（人）')
    axes[i].grid(axis='both')
    axes[i].legend()

# ---------------------------------------------
    
# 添加总标题
fig.suptitle('两影院三类影片平均票价与平均上座分布', fontsize=20, y=1.05)

# 展示绘图结果
plt.tight_layout()
plt.show()
```
