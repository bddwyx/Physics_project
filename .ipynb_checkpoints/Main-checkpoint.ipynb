```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

# 初速度恒定的磁聚焦现象 

## 磁聚焦原理

&nbsp; 在Z轴沿Z轴正方向加上一个大小恒定的磁场。带电量为q的带电粒子初速度与Z轴有一个恒定夹角$\theta$，则对于该粒子，其运动轨迹在X-Y平面上投影为圆周运动，半径满足
$$R = \frac {mv\sin\theta} {Bq}$$

&nbsp; 其螺距满足
$$d = \frac {2\pi mv \cos \theta} {Bq}$$

&nbsp; 运动轨迹为
$$x(t) = 2R(1 - \cos{\frac{Bq}{m}t})$$
$$y(t) = R\sin{\frac{Bq}{m}t}$$
$$z(t) = v\cos\theta t$$

## 磁聚焦空间轨迹图

```python
#constant
B = 1
q = 1
m = 1
v = 1

t = np.linspace(0, 30, 200)
```

```python
def getData(theta):
    #equavalent
    R = m * v * np.sin(theta) / B / q
    w = B * q / m

    x = 2 * R * (1 - np.cos(w * t))
    y = R * np.sin(w * t)
    z = v * np.cos(theta) * t
    
    return x, y, z
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

fig = plt.figure(figsize = (12, 5))
ax = fig.add_subplot(1, 2, 1,projection='3d')
ay = fig.add_subplot(1, 2, 2,projection='3d')

for i in range(1,4):
    x, y, z = getData(i / 200)
    ax.plot(x, y, z, label='curve' + str(i))
    ay.plot(x, -z, y, label='curve' + str(i))

ax.legend()
ax.set_title('Z轴向上观测图')
ax.set_xlabel('x/m')
ax.set_ylabel('y/m')
ax.set_zlabel('z/m')

ay.legend()
ay.set_title('轨迹侧向观测图')
ay.set_xlabel('x/m')
ay.set_ylabel('z/m')
ay.set_zlabel('y/m')

fig.suptitle("磁聚焦现象空间轨迹图", fontsize = 20, y = 1.05)
plt.tight_layout()
plt.show()
```

```python
fig2 = plt.figure(figsize = (10, 5))
ap1 = fig2.add_subplot(1, 2, 1)
ap2 = fig2.add_subplot(1, 2, 2)
for i in range(1,4):
    x, y, z = getData(i / 200)
    ap1.plot(x, y, label='curve' + str(i))
    ap2.plot(z, y, label='curve' + str(i))
    
ap1.set_title("运动轨迹在X-Y平面投影")
ap1.set_xlabel('x/m')
ap1.set_ylabel('y/m')
ap1.legend(loc = 'upper right')

ap2.set_title("运动轨迹在Y-Z平面投影")
ap2.set_xlabel('z/m')
ap2.set_ylabel('y/m')
ap2.legend(loc = 'upper right')

fig2.suptitle("空间轨迹投影图", fontsize = 20, y = 1.05)
plt.tight_layout()
plt.show()
```
